##### Krill Herd

##### Import Libraries

```python
import numpy as np
import random 
import matplotlib.pyplot as plt
import math
import cv2
import scipy.special as sp
```

```python
def initkrillherd(num_krillherd,dim,p_high,p_low):
    krillherd_pos=[]
    while len(krillherd_pos)<num_krillherd:
        krillherd_pos.append([p_low+(random.uniform(0, 1)*(p_high-p_low)) for _ in range(dim)])
    krillherd_pos = list(krillherd_pos)
    return krillherd_pos
```

```python
def fitness_function(x):
    return 1/(1+(x[0]**2+x[1]**2+x[2]**2))
```

```python
def visualize_krillherd(positions, video_filename='krillherd.mp4'):
    fig, ax = plt.subplots(figsize=(10, 7))
    out = cv2.VideoWriter(video_filename, cv2.VideoWriter_fourcc(*'XVID'), 10, (640, 480))

    for gen, pos in enumerate(positions):
        print(pos)
        ax.cla()
        pos_x = [p[0] for p in pos]
        print(pos_x)
        pos_y = [p[1] for p in pos]
        print()
        ax.scatter(pos_x, pos_y, c='blue', label='krillherd')
        print(gen)
        #ax.scatter(posbest[gen][0], posbest[gen][1], c='red', label='Best cat', marker='*')
        ax.set_title(f'Iteration {gen + 1}')
        ax.set_xlim(-20, 20)
        ax.set_ylim(-20, 20)
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.legend()
        # Save the frame to a video
        fig.canvas.draw()
        img = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
        img = img.reshape(fig.canvas.get_width_height()[::-1] + (3,))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        img = cv2.resize(img, (640, 480))  # Resize to fit the video frame
        out.write(img)

    out.release()
    plt.close()
```

```python
num_krill=100
num_iter=200
dim=3
p_high=5
p_low=-5
vf=0.02
rdmax=0.6
pos_krillherd=initkrillherd(num_krill,dim,p_high,p_low)
fitness=[]
for i in pos_krillherd:
    fitness.append(fitness_function(i))
positions=[[list(krill) for krill in pos_krillherd]]
```

```python
iter=0
theetamain=[]
foragingmain=[]
while iter<num_iter:
    theeta=[]
    foraging=[]
    diffusion=[]
    for i in range(num_krill):
        temp=2*(random.uniform(0,1)+iter/num_iter)*max(fitness)
        index=fitness.index(max(fitness))
        e_target=[temp*pos_krillherd[index][di] for di in range(dim)]
        temp1=(1/(5*num_krill))
        sd=np.zeros(dim).tolist()
        for j in range(num_krill):
            for k in range(dim):
                sd[k]=abs(pos_krillherd[j][k]-pos_krillherd[i][k])
        for j in range(dim):
            sd[j]*=temp1
            #if(j==2):
                #print(sd[j])
        e_local=np.zeros(dim).tolist()
        for j in range(num_krill):
            for k in range(dim):
                if(abs(pos_krillherd[j][k]-pos_krillherd[i][k])<sd[k] and i!=j):
                    a=(abs(pos_krillherd[j][k]-pos_krillherd[i][k]))*random.uniform(0.00001,1)
                    if(a==0):
                        xij=0.00000002
                    else:
                        xij=pos_krillherd[j][k]-pos_krillherd[i][k]
                    fij=fitness[j]-fitness[i]/min(fitness)-max(fitness)
                    e_local[k]+=fij*xij
        ei=[e_local[di]+e_target[di] for di in range(dim)]
        nun=random.uniform(0.4,0.9)
        theetamax=random.uniform(np.pi/6, np.pi/2)
        if len(theetamain)==0:
            theeta.append([ei[di]*theetamax for di in range(dim)])
        else:
            theeta.append([ei[di]*theetamax+nun*theetamain[-1][i][di] for di in range(dim)])
        ai=fitness[i]+max(fitness)
        if(len(foragingmain)<1):
            foraging.append([vf*ai for di in range(dim)])
        else:
            foraging.append([vf*ai+random.uniform(0,1)*foragingmain[-1][i][di] for di in range(dim)])
        diffusion=[rdmax*random.uniform(-1,1) for di in range(dim)]
        dxdt=[theeta[i][di]+foraging[i][di]+diffusion[di] for di in range(dim)]
        dt=random.uniform(-1,1)*(p_high-p_low)*dim
        xnew=[pos_krillherd[i][di]+dt*dxdt[di] for di in range(dim)]
        for di in range(dim):
            mr=0.05/max(fitness)
            cr=0.2*max(fitness)
            if(random.uniform(0,1)<cr):
                r=random.choice([j for j in range(num_krill) if i!=j])
                xnew[di]=(1-cr)*pos_krillherd[i][di]+cr*pos_krillherd[r][di]
            if(random.uniform(0,1)<mr):
                r1=random.choice([j for j in range(num_krill) if i!=j])
                r2=random.choice([j for j in range(num_krill) if i!=j and j!=r1])
                xnew[di]= pos_krillherd[fitness.index(max(fitness))][di]+random.uniform(0,1)*(pos_krillherd[r1][di]-pos_krillherd[r2][di])
        if(fitness_function(xnew)>min(fitness)):
            print("yes")
            index=fitness.index(min(fitness))
            fitness[index]=fitness_function(xnew)
            pos_krillherd[index]=xnew
    positions.append([list(krill) for krill in pos_krillherd])
    theetamain.append(theeta)
    foragingmain.append(foraging)
    iter+=1
```

```python
visualize_krillherd(positions)
```
