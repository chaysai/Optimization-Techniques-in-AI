##### Cuckoo Search Algorithm

##### import libraries

```python
import numpy as np
import random 
import matplotlib.pyplot as plt
import math
import cv2
import scipy.special as sp
```

##### initialise cuckoo positions

```python
def initialise_cuckoo(num_cuckoo,dim):
    cuckoo= set()
    while len(cuckoo) < num_cuckoo:
        temp = tuple(random.uniform(0, 5) for _ in range(dim))
        cuckoo.add(temp)
    
    cuckoo= list(cuckoo)
    return cuckoo
```

#####  calculate fitness function

```python
def fitness_function(x):
    return x[0]**2+x[1]**2
```

##### Find best

```python
def find_best(a,x):
    f_best=a[a.index(min(a))]
    x_best=x[a.index(min(a))]
    return f_best,x_best
```

##### calculate sigma u

```python
def calculate_sigma_u(beta):
    gamma_1_plus_beta = sp.gamma(1 + beta)
    sin_term = math.sin(math.pi * beta / 2)
    gamma_half_1_plus_beta = sp.gamma((1 + beta) / 2)
    beta_term = beta ** (2 * (1 - beta) / 2)
    
    sigma_u = (gamma_1_plus_beta * sin_term / (gamma_half_1_plus_beta * beta_term)) ** (1 / beta)
    return sigma_u
```

##### visualisation and video creation

```python
def visualize_cuckoo(positions, fitbest, posbest, video_filename='cuckoo_search_algorithm.mp4'):
    fig, ax = plt.subplots(figsize=(10, 7))
    out = cv2.VideoWriter(video_filename, cv2.VideoWriter_fourcc(*'XVID'), 10, (640, 480))

    for gen, pos in enumerate(positions):
        ax.cla()
        pos_x = [p[0] for p in pos]
        pos_y = [p[1] for p in pos]
        ax.scatter(pos_x, pos_y, c='blue', label='cuckoo')
        print(gen)
        ax.scatter(posbest[gen][0], posbest[gen][1], c='red', label='Best cuckoo', marker='*')
        ax.set_title(f'Iteration {gen + 1}')
        ax.set_xlim(-5, 5)
        ax.set_ylim(-5, 5)
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.legend()

        # Save the frame to a video
        fig.canvas.draw()
        img = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
        img = img.reshape(fig.canvas.get_width_height()[::-1] + (3,))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        img = cv2.resize(img, (640, 480))  # Resize to fit the video frame
        out.write(img)

    out.release()
    plt.close()
```

```python
n_cuckoo=100
pos_cuckoo=initialise_cuckoo(n_cuckoo,2)
positions=[]
fitbest=[]
posbest=[]
all_fitness=[]

positions=[[list(cuckoo) for cuckoo in pos_cuckoo]]
fitness=[]
for i in pos_cuckoo:
    fitness.append(fitness_function(i))
all_fitness.append(fitness)
f_best,cuckoo_best=find_best(fitness,pos_cuckoo)
fitbest.append(f_best)
posbest.append(cuckoo_best)
iter=600
```

```python
gen=0
beta=1.5
while gen<iter:
    for i in range(n_cuckoo):
        u = np.random.normal(0, calculate_sigma_u(beta),1)
        v=np.random.normal(0,1,1)
        s=u/((abs(v))**(1/beta))
        xnew=[(pos_cuckoo[i][dim]+random.uniform(0,1)*0.01*s*(pos_cuckoo[i][dim]-cuckoo_best[dim]))[0] for dim in range(2)]
        print(xnew)
        if(fitness_function(xnew)<fitness_function(pos_cuckoo[i])):
            pos_cuckoo[i]=xnew
    fitness=[]
    for i in pos_cuckoo:
        fitness.append(fitness_function(i))
    all_fitness.append(fitness)
    f_best,cuckoo_best=find_best(fitness,pos_cuckoo)
    fitbest.append(f_best)
    posbest.append(cuckoo_best)
    positions.append([list(cuckoo) for cuckoo in pos_cuckoo])
    gen+=1


```

```python
visualize_cuckoo(positions, fitbest, posbest)
```
